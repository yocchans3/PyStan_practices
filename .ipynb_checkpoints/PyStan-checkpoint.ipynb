# Bayes 更新

```python
p_a = 3.0/10.0
p_b = 5.0/10.0
p_prior = 0.5 #袋aを選ぶ確率
#0:blu, 1:red
data = [0, 1, 0, 0, 1, 1, 1]
```

```python
N_data = 7 #ベイズ更新でどこまでのデータを使用するか
likelihood_a = bernoulli.pmf(data[:N_data], p_a)
likelihood_b = bernoulli.pmf(data[:N_data], p_b)
```

```python
likelihood_a
```

```python
pa_posterior = p_prior
pb_posterior = p_prior
pa_posterior *= np.prod(likelihood_a) #事後分布
pb_posterior *= np.prod(likelihood_b) #事後分布

#規格化normalization
norm = pa_posterior + pb_posterior

df = pd.DataFrame([pa_posterior/norm, pb_posterior/norm], columns=["post"])
x = np.arange(df.shape[0])
plt.bar(x, df["post"])
plt.xticks(x, ["a", "b"])
```

# 棄却サンプリング

```python
from scipy import optimize as opt
from scipy.stats import beta, uniform
```

```python
np.random.seed(123)
```

```python
a, b = 1.5, 2.0
x = np.linspace(beta.ppf(0.001, a, b), beta.ppf(0.999, a, b), 100)
plt.plot(x, beta.pdf(x, a, b))
```

```python
f = beta(a=a, b=b).pdf
res = opt.fmin(lambda x:-f(x), 0.3)
y_max = f(res)
```

```python
y_max
```

```python
NMCS = 5000
x_mcs = uniform.rvs(size = NMCS)
r = uniform.rvs(size = NMCS)*y_max
accept = x_mcs[r <= f(x_mcs)]
plt.hist(accept, density=True, bins=30, rwidth=0.8, label="rejection sampling")
plt.plot(x, beta.pdf(x, a, b), label = "Target distribution")
plt.legend()
```

# 次元の呪い

```python
from scipy.special import gamma
```

```python
accept_dict = {}
```

```python
NMCS = 40000
ndim = 16
accept = 0
for i in range(NMCS):
    x = 2*np.random.rand(ndim)-1.0
    r = (np.sum(x**2))**0.5
    if r <=1:
        accept += 1
accept_ratio = accept/NMCS
analytical = np.pi**(ndim/2)/(2**ndim*gamma(ndim/2+1))
print("accept ratio:", accept_ratio)
print("analytical solution", analytical)
accept_dict.update({ndim:accept_ratio/analytical})
```

```python
accept_dict
```

```python
df = pd.DataFrame.from_dict(accept_dict, orient="index")
```

```python
plt.scatter(df.index, df[0])
plt.xlabel("dimension")
plt.ylabel("ratio [-]")
```

# (ランダムウォーク)M-Hアルゴリズム

```python
np.random.seed(123)
```

```python
a, b = 1.5, 2.5
x = np.linspace(beta.ppf(0.001, a, b), beta.ppf(0.999, a, b), 100)
plt.plot(x, beta.pdf(x, a, b))
```

```python
theta = 0.8
NMCS = 20000
epsilon = 0.5
theta_mcs = [theta]
for i in range(NMCS):
    theta_new = theta + epsilon * np.random.randn()
    if beta.pdf(theta_new, a, b) > beta.pdf(theta, a, b):
        theta = theta_new
    else:
        r = beta.pdf(theta_new, a, b) / beta.pdf(theta, a, b)
        if np.random.rand() < r:
            theta = theta_new
    theta_mcs.append(theta)
df = pd.DataFrame(theta_mcs)
```

```python
df.head()
```

```python
plt.plot(df[0])
plt.xlabel("MCS")
plt.ylabel("$\Theta$")
```

```python
plt.hist(df[0][1000:], density=True, bins=40)
plt.plot(x, beta.pdf(x, a, b))
```

# Gibbs sampler

```python
from scipy.stats import norm, multivariate_normal
```

```python
from IPython.html.widgets import interact
```

```python
mu = np.array([1, 2])
cov = np.array([[1.0, 0.7], [0.7, 1.0]])
```

```python
x = np.arange(-2, 4, 0.01)
y = np.arange(-1, 5, 0.01)
X, Y = np.meshgrid(x, y)
pos = np.zeros([X.shape[0], X.shape[1], 2])
pos[:, :, 0] = X
pos[:, :, 1] = Y
```

```python
rv = multivariate_normal(mu, cov)
plt.figure(figsize = (6,6))
plt.contour(X, Y, rv.pdf(pos), cmap="plasma")
plt.xlabel("$x_{1}$")
plt.ylabel("$x_{2}$")
```

```python
NMCS = 500
sigma12 = cov[1, 0]
sigma1 = cov[0, 0] ** 0.5
sigma2 = cov[1, 1] ** 0.5
rho = sigma12 / (sigma1 * sigma2)
z1 = 0
z2 = 0
z1_mcs = [z1]
z2_mcs = [z2]
for i in range(NMCS):
    #f(z1|z2)
    z1 = norm.rvs(loc=rho*z2, scale=(1-rho**2)**0.5)
    z1_mcs.append(z1)
    z2_mcs.append(z2)
    #f(z2|z1)
    z2 = norm.rvs(loc=rho*z1, scale=(1-rho**2)**0.5)
    z1_mcs.append(z1)
    z2_mcs.append(z2)
df1 = pd.DataFrame(mu[0]+np.array(z1_mcs)*sigma1)
df2 = pd.DataFrame(mu[1]+np.array(z2_mcs)*sigma2)
```

```python
df1.head()
```

```python
plt.plot(df1[0], df2[0])
```

```python
#labだとアニメーションできんわ、、、
@interact(mcs=(0, 100, 1))
def animation(mcs=0):
    plt.contour(X, Y, rv.pdf(pos), cmap = "plasma")
    plt.xlim([-3, 4.5])
    plt.ylim([-1, 5])
    plt.plot(df1[0][:mcs], df2[0][:mcs])
    plt.show()
```
